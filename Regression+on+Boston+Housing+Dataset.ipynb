```python
from sklearn import datasets
boston=datasets.load_boston()
```

```python
import numpy as np
import pandas as pd
```

```python
y=pd.DataFrame(data= boston['target'])
y.head()
```

```python
x=pd.DataFrame(data= boston['data'])
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
X_rooms=x.iloc[:, 0]
X_rooms=np.array(X_rooms)
X_rooms=X_rooms.reshape(-1,1)
X_rooms
```

```python
y.shape
```

```python
y=np.array(y)
y=y.reshape(-1,1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_rooms, y,color='darkorange', label='data')
plt.ylabel('Value of house /1000 ($)') 
plt.xlabel('Number of rooms')
```

```python
from numpy import linspace
from sklearn import linear_model 
```

```python
alg = linear_model.LinearRegression() 
alg.fit(X_rooms, y)
k=linspace(min(X_rooms),max(X_rooms)).reshape(-1,1) 
plt.scatter(X_rooms, y, color='blue') 
plt.plot(k, alg.predict(k),'b', linewidth=3) 
```

```python
 from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(x, y,test_size = 0.3, random_state=42)
```

```python
alg2 = linear_model.LinearRegression() 
alg2.fit(X_train, y_train) 
y_pred = alg2.predict(X_test)
alg2.score(X_test, y_test)
```

```python
from sklearn.linear_model import Lasso
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size = 0.3, random_state=42) 
lasso = Lasso(alpha=0.1, normalize=True)
lasso.fit(X_train, y_train)
lasso_pred = lasso.predict(X_test)
lasso.score(X_test, y_test)
```

```python
from sklearn.linear_model import Ridge
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size = 0.3, random_state=42)
ridge = Ridge(alpha=0.1, normalize=True) 
ridge.fit(X_train, y_train) 
ridge_pred = ridge.predict(X_test) 
ridge.score(X_test, y_test)
```
