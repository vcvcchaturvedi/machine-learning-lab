```python
# Import packages 
import numpy as np
from sklearn import decomposition, datasets 
from sklearn.preprocessing import StandardScaler
```

```python
# Load the breast cancer dataset 
dataset = datasets.load_breast_cancer()
# Load the features 
X = dataset.data
```

```python
# View the shape of the dataset 
X.shape
```

```python
# Create a scaler object 
sc = StandardScaler()

# Fit the scaler to the features and transform 
X_std = sc.fit_transform(X)
```

```python
# Create a pca object with the 2 components as a parameter 
pca = decomposition.PCA(n_components=3)
# Fit the PCA and transform the data 
X_std_pca = pca.fit_transform(X_std)
```

```python
# View the new feature data's shape 
X_std_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
 from sklearn.cluster import KMeans
```

```python
 # Load data
iris = datasets.load_iris() 
X = iris.data
```

```python
# Standarize features 
scaler = StandardScaler() 
X_std = scaler.fit_transform(X)

# Create k-mean object 
clt = KMeans(n_clusters=3, random_state=0, n_jobs=-1) 
```

```python
# Train model
model = clt.fit(X_std) 
```

```python
# View predict class 
model.labels_
```

```python
# Create new observation 
new_observation = [[0.8, 0.8, 0.8, 0.8]]
```

```python
# Predict observation's cluster 
model.predict(new_observation)[0]
```

```python

# View cluster centers 
model.cluster_centers_
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns # nice color
from sklearn import cluster
```

```python
iris = datasets.load_iris() 
X = iris.data[:, :2] # use only 'sepal length and sepal width' 
y_iris = iris.target
```

```python
km2 = cluster.KMeans(n_clusters=2).fit(X) 
km3 = cluster.KMeans(n_clusters=3).fit(X) 
km4 = cluster.KMeans(n_clusters=4).fit(X)
```

```python
plt.figure(figsize=(5,15)) 
plt.subplot(311) 
plt.scatter(X[:, 0], X[:, 1], c=km2.labels_) 
plt.title("K=2, J=%.2f" % km2.inertia_) 
plt.subplot(312)
plt.scatter(X[:, 0], X[:, 1], c=km3.labels_) 
plt.title("K=3, J=%.2f" % km3.inertia_) 
plt.subplot(313)
plt.scatter(X[:, 0], X[:, 1], c=km4.labels_)#.astype(np.float)) 
plt.title("K=4, J=%.2f" % km4.inertia_)
```
